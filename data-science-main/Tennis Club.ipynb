```python
from geopy.distance import distance
def get_distance(df1, df2):
    df_distance = pd.DataFrame(columns = ['Neighborhood','ID','ADDRESS','LAT','LONG','Distance to Venue'])
    i=0
    for lat_n, lng_n, fsa in zip(df2['latitude'], df2['longitude'], df2['Neighborhood']):
        for lat_v, lng_v, id_v, name_v in zip(df1['LAT'], df1['LONG'], df1['ID'], df1['ADDRESS']):
            i = i + 1
            dv = distance([lat_n, lng_n], [lat_v, lng_v]).m
            df_distance.loc[i] = [fsa]+[id_v]+[name_v]+[lat_v]+[lng_v] + [dv]
    return df_distance
```

```python
from sklearn.cluster import KMeans
sse = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k).fit(X_train)
    kmeans.fit(X_train)
    sse.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
plt.plot(range(1, 15), sse)
plt.xticks(range(1, 15))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```
